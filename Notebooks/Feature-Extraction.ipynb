we will consider the cities where minimum 5 maches are played

```python
city_counts = df['city'].value_counts()
cities_to_keep = city_counts[city_counts > 600].index.tolist()
df = df[df['city'].isin(cities_to_keep)]
```

```python
df['city'].value_counts()
```

## Caluculating the total runs

```python
df.info()
```

```python
df['runs'] = pd.to_numeric(df['runs'])
df['current_score'] = df.groupby('match_id')['runs'].cumsum()
```

```python
df
```

```python
df.info()
```

## Balls and Wickets Left

```python
num=15.4
left = str(num).split('.')[0]
right = str(num).split('.')[1]
left=int(left)
right=int(right)
```

```python
df['left'] = df['ball'].apply(lambda x:str(x).split('.')[0]).astype('int')
df['right'] = df['ball'].apply(lambda x:str(x).split('.')[1]).astype('int')
```

```python
df['ballls_bowled']=(df['left']*6)+df['right']
df['balls_left']=120-df['ballls_bowled']
df['balls_left']=df['balls_left'].apply(lambda x:0 if x<0 else x)
```

```python
df['player_dismissed'] = df['player_dismissed'].apply(lambda x:0 if x=='0' else 1)
df['player_dismissed'] = df['player_dismissed'].astype('int')
df['player_dismissed'] = df.groupby('match_id')['player_dismissed'].cumsum()
df['wickets_left'] = 10 - df['player_dismissed']
```

```python
df
```

## Current Run Rate

```python
df['crr']=df['current_score']/(df['ballls_bowled']/6)
```

```python
df
```

```python
df = df.rename(columns={'ballls_bowled': 'balls_bowled'})
```

```python
df
```

```python
last_five = []

for match_id, group in df.groupby('match_id'):
    group = group.sort_values('balls_bowled').reset_index(drop=True)
    match_list = []
    for id in range(len(group)):
        if id < 30:
            match_list.append(np.nan)
        else:
            sum_last_30 = group.loc[id-30:id-1, 'runs'].sum()
            match_list.append(sum_last_30)

    last_five.extend(match_list)
```

```python
df['last_five'] = last_five
```

```python
df
```

## Total Runs - Target Col

```python
final_df = df.groupby('match_id').sum()['runs'].reset_index().merge(df,on='match_id')
```
